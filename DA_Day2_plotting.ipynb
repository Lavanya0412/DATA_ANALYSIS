**problem-1**

Runs scored by 10 players [100,50,91,78,89,25,34,19,9,10] wickets taken by same 10 new players [1,0,2,0,3,7,8,9,7,5]

Form clusters for batsmen and bowlers

```python
#Runs scored by 10 players [100,50,91,78,89,25,34,19,9,10] wickets taken by same 10 new players [1,0,2,0,3,7,8,9,7,5]
import numpy as np
runs=np.array([100,50,91,78,89,25,34,19,9,10]
wickets=np.array([1,0,2,0,3,7,8,9,7,5]
plt.scatter(runs,wickets)
plt.show()
```

```python
#coloring
import numpy as np
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.plot(runs,wickets,color='red')
plt.show()
```

```python
runs=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color='r',marker="p")
plt.show()
```

**problem-2**

plot the scores of two students in five different subjects.Subjects as x axis and marks as y axis

```python
std1=np.array([76,98,92,95,78])
std2=np.array([77,89,98,73,65])
sub=["maths","science","english","telugu","social"]
plt.plot(sub,std1,label="student-1",linestyle=':')
plt.plot(sub,std2,label="student-2")
plt.legend()
```

```python
std1=np.array([76,98,92,95,78])
std2=np.array([77,89,98,73,65])
std3=np.array([98,78,45,68,90])
sub=["maths","science","english","telugu","social"]
plt.plot(sub,std1,label="student-1")
plt.plot(sub,std2,label="student-2")
plt.plot(sub,std3,label="studdent-3")
plt.legend()
```

```python
std1=np.array([76,98,92,95,78])
std2=np.array([77,89,98,73,65])
sub=["maths","science","english","telugu","social"]
plt.subplot(2,1,1)
plt.plot(sub,std1,label="student-1",color="b")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,std2,label="student-2",color="k")
plt.legend()
```

create subplots(bar) showing the profits of two companies
Given:

year      A       B
2019     ( 230     200)
2020      (560     160)
2021       (780     270)
2022      (127      127)
2023      (128      400)

```python
comp1=np.array([230,560,780,127,128])
comp2=np.array([200,160,270,127,400])
print("Revenue of company 1",comp1)
print("Revenue of company 2",comp2)
years=["19-20","20-21","21-22","22-23",]
prof1=np.array(np.diff(comp1))
print("Profits of company 1",prof1)
prof2=np.array(np.diff(comp2))
print("Profits of company 2",prof2)
plt.subplot(1,2,1)
plt.bar(years,prof1,label="profits-1",color="k")
plt.legend()
plt.subplot(1,2,2)
plt.bar(years,prof2,label="profits-2",color="g")
plt.legend()




```
